## Grafico de Dispersion (Scatter Plot)

```python
import numpy as np
import matplotlib.pyplot as plt

# Preparacin de datos
N = 1500 #Base de 1500 puntos
X = np.random.randn(N) #Eje Random de 1500 puntos
Y = 2.5*X+np.random.randn(N) #Eje Y multilicamos por 2.5 los valores de X y le sumamos el random de X genera dispersion.

#grafico
plt.scatter(X,Y,alpha=0.4) #Tipo de grafico, posicion de los ejes y el alpha que es el nivel de transparencia
plt.title("Grafico de dispersion con Matplotlib")
plt.show()
```

## Grafico de Lineas (Line Plot)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Preparacin de datos
ventas = [1200, 1500, 1700, 1600, 2000,2100]
meses = ['Enero','Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']

#Graficos de lineas con Seaborn
df = pd.DataFrame({'Meses':meses, 'Ventas':ventas}) #Creamos el dataframe con los meses y las ventas
sns.lineplot(data=df, x="Meses", y="Ventas", marker="o", color="green", sort=False, lw=1) #indicamos los ejes X,Y, indicamos el tipo de marcador y el color de linea
plt.title("Ventas Mensuales 2024")
plt.show()
```

## Grafico de histograma

```python
import matplotlib.pyplot as plt
import numpy as np

#Generamos datos, LOC es la media donde se ubicaran los "datos", "scale" es la desviacion estandar y size el numero de datos a crear
datos = np.random.normal(loc=50, scale=10, size=1000) 

#Creamos el grafico, "bins" es el numero d barras, purpura y bordes negros
plt.hist(datos, bins=30, color="purple", edgecolor="black") 
plt.title("Distribucin de calificaciones")
plt.xlabel("puntaje") #label x
plt.ylabel("Frecuencia") #label y
plt.show()
```

## Grafico de Barras

```python
#El Grafico de barras, no cuenta cuantas veces aparece una varible, si no que representa el valor de esa variable
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

categorias = ['A', 'B','C','D']
valores = [23,45,56,78]

df = pd.DataFrame({"Categorias": categorias, "Valores": valores})
#Asignamos las categoras, paleta de colores, el "hue" es que se coloree de acuerdo a las categorias, "dodge" es para agrupar las barras
sns.barplot(x="Categorias", y="Valores", data = df, order=df.sort_values('Valores').Categorias, orient="x", palette="Blues_d", hue="Categorias", dodge=False)  
plt.title("Comparacin de Categorias")
plt.legend([],[], frameon=False) # Ocultamos las leyendas
plt.show()
```

## Grafico de Cajas

```python
import pandas as pd
import seaborn as sns

data = {
    'Categoria 1': np.random.normal(100,10,200), #Media 100, devicion 10, 200 datos
    'Categoria 2': np.random.normal(80,20,200),
    'Categoria 3': np.random.normal(90,15,200)
}
df = pd.DataFrame(data) # LO volvemos DataFrame

sns.boxplot(data=df, palette="Set3")
plt.title("Ditribucin por categorias")
plt.show()
```

## Grafico Circular (Pie Chart)

```python
import matplotlib.pyplot as plt

tamanios = [50,25,15,10]
etiquetas = ['A','B','C','D']

#Con autopct indicamos cuantos decimales y en formato de % , startangle es desde que angulo se va a empezar a graficar
plt.pie(tamanios, labels=etiquetas, autopct='%1.1f%%', startangle=140) 
plt.title("Distribucin de particiapacion")
plt.show()
```

## Mapa de Calor (Heat Map)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

dias_Semana = ['Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domimgo']
#Creamos un rango de horario entre las 6 y las 5 de la tarde
horas_Dia= [f"{h}:00" for h in range(6,18)] 
#Creamos el ramdon con 12 filas(horas) y 7 columnas(dias), el indice seran las horas del dia, media 0, desviacion estandar 1
calificaciones = pd.DataFrame(np.random.randn(12,7), index=horas_Dia, columns=dias_Semana)

#Creamos el grafico, "RdYlGn" es la escala de colores y "Annot" es si los numero aparecen o no
sns.heatmap(calificaciones, cmap="RdYlGn", annot=True)
plt.title("Mapa de calor, calificaciones por hora") #Titulo
plt.xlabel("Dias de la semana") #Label eje X
plt.ylabel("Dias de la semana") #Label eje Y
plt.show()
```

## Grafico de Araas (Spider Graph)

```python
import matplotlib.pyplot as plt
import numpy as np

labels = np.array(['Velocidad', 'Fuerza', 'Resistencia', 'Inteligencia', 'Magia'])
valores= [7,8,6,9,7]

#grafico de araa, hacemos el distanciamiento
# linspace indicamos que los circulos se generan de 0 a 2 radianes por PI, len(labels) es el numero de diviciones, endpoint es para finalizar,
# al terminar de graficar en 360 y todo lo convertimos en una lista
angulos = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
valores += valores[:1] #Tomamos el primer valor de la lista valores y lo ponemos al final
angulos += angulos[:1] #Tomamos el primer valor de la lista angulos y lo ponemos al final

#Creamos una imagen de 6 x 6 pulgadas y un grafico polar
fig, ax= plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
ax.fill(angulos, valores, color='red', alpha=0.25) #Rellenamos el grafico y con el alpha aplicamos degrado
ax.plot(angulos, valores, color='red', linewidth=2) #Delineamos
ax.set_yticklabels([]) #Eliminamos las etiquetas del eje Y
ax.set_xticks(angulos[:-1]) #Use todos los angulos menos el ultimo, ya que lo habiamos pasado de primeras al final
ax.set_xticklabels(labels) #Usamos la etiquetas del eje X

plt.title("Perfil de un personaje")
plt.show()
```

## Grafico de pares (Pair Plot)

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Origen de datos, MPG es una ejemplo ya precargado de SeaBorn
mpg = sns.load_dataset('mpg')

#Seleccionamos algunas variables tipos de datos del dataset, por cada variable, se crea un grafico
subset = mpg[['mpg', 'horsepower', 'weight', 'acceleration', 'origin']]

#Creamos el grafico, usamos las subset como etiquetas, "hue" indica que cada dato se colorea diferente segun el tipo de origen,
#la paleta de colores va a ser SET1
sns.pairplot(subset, hue='origin', palette='Set1')
#Titulo para todo el grafico, la variable "Y" indica que la distancia de texto titulo respecto a los graficos
plt.suptitle("Grafico de pares para el DataSet de MPG, Agrupado por Origen", y=1.02)
plt.show()
```
